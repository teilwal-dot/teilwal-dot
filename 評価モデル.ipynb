```python
import gensim

# 学習済モデルのパス
model_path = '/content/drive/MyDrive/特別研究/crawl-300d-2M.vec'
# ロードに10分くらいかかる
model = gensim.models.KeyedVectors.load_word2vec_format(model_path, binary=False)
```

```python
# 登録している単語の数
print(len(model.vocab.keys())) # 2000000
```

```python
# ひとつの単語ベクトルの次元
model['cat'].shape # (300,)
```

```python
#コサイン類似度が近いものトップ10
from pprint import pprint

pprint(model.most_similar('cat', topn=10))
```

```python
# コサイン類似度の計算
print(model.similarity('one', 'people'))
print(model.similarity('one', 'dark'))
print(model.similarity('one', 'black'))
print(model.similarity('one', 'bike'))
```
